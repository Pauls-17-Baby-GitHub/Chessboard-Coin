**PROBABILITY OF COIN FALLING ON CHESSBOARD FALL ON BOTH BLACK AND WHITE**

```
import random
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```
event = 0                                                                       # positive event count variable
numerical_probability = []
throws = 0                                                                      # total event count variable

# circle properties
coin_color = (0, 0, 255)                                                        # Red (B, G, R)
center_color = (0,255,0)                                                        # Green
thickness = 2                                                                   # in pixels
radius  = 500                                                                   # scaled radius to 0.5 units

image = cv2.imread('/content/square.png')                                       # read chessboard image

# new_size
new_size = (2000, 2000)

output = cv2.resize(image, new_size, interpolation = cv2.INTER_AREA)            # resize image to 2000 x 2000 to scale to 2 unit length square

n = 1000                                                                        # number of samples taken
random.seed(19)
for i in range(0,n):
  throws += 1
  x = random.uniform(0,2)
  y = random.uniform(0,2)
  center_coordinates = (int(x*1000),int(y*1000))                                # scaled center coordinate
  output = cv2.circle(output, center_coordinates, 5, center_color, -1)          # draw centre point
  output = cv2.circle(output, center_coordinates, radius, coin_color, thickness)# draw coin
  #print('(', x , ',', y , ')')                                                 # actual center coordinate
  if x > 1.5 or x < 0.5 or y > 1.5 or y < 0.5:                                  # if out of the inner square, then overlaps two colors
      event += 1
  numerical_probability.append(event/throws)                                    # probability = favorable_events_occurred/total_events_occurred

print(numerical_probability)

  
cv2_imshow(output)                                                              # Displaying the 2x2 chessboard image


plt.plot(numerical_probability)                                                 # plot
plt.xlabel("Throw count")
plt.ylabel("Numerical probability")
```
